```python
import numpy as np
import matplotlib.pyplot as plt

X0 = 1
T = 1.0
N = 2**8
dt = T / N

alpha = 1
sigma = 2

t = np.arange(0.0, T, dt)
dW = np.sqrt(dt) * np.random.randn(N)
dW[0] = 0
W = np.cumsum(dW)

X_numerical = np.zeros(N)
Xt = X0
X_numerical[0] = X0
for j in range(1, N):
    Xt = Xt - alpha*Xt*dt + sigma*dW[j] 
    X_numerical[j] = Xt

X_analytical = np.zeros(N)
Xt_a = X0
Integral = 0
for k in range(0, N-1):
    Integral = Integral + np.exp(alpha*dt)*dW[k]
    Xt_a = np.exp(-alpha*(k/N))*X0 + sigma*np.exp(-alpha*(k/N)) * Integral 
    X_analytical[k] = Xt_a

plt.plot(t, X_analytical, label='Analytical-Numerical')
plt.plot(t, X_numerical, label='Numerical')
plt.legend()
```
